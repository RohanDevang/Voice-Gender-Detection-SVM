### **2. Data Preprocessing**
- We are working with **Librosa** (a popular Python library for audio analysis), converting **M4A to WAV** is necessary due to the following reasons:


**1. Librosa Does Not Natively Support M4A**
- **Librosa** primarily uses `soundfile`for loading audio files.
- `soundfile` supports **WAV, FLAC**, and a few others but **not M4A**.

**2. WAV is Uncompressed, Ensuring Accurate Audio Processing**
- **Librosa loads audio as NumPy arrays** for analysis (e.g., computing spectrograms, MFCCs, etc.).
- M4A is a **compressed format**, meaning some audio information might be lost.
- WAV provides **raw, lossless audio**, ensuring better processing accuracy.

**3. Avoiding Compatibility Issues with Feature Extraction**
- Some **Librosa functions may fail with compressed formats** like M4A.
- WAV files are preferred for tasks like:
  - **Pitch detection (`librosa.piptrack()`)**
  - **Onset detection (`librosa.onset_detect()`)**
  - **Mel spectrograms (`librosa.feature.melspectrogram()`)**

**4. Consistency in Sample Rate Handling**
- Librosa **automatically resamples audio** to **22,050 Hz** by default.
- With **compressed formats like M4A, resampling may introduce errors**.
- Using WAV ensures **consistent sample rate conversion** without artifacts.

```python
from pydub import AudioSegment
import os

def convert_m4a_to_wav(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    
    for file in os.listdir(input_folder):
        if file.endswith(".m4a"):
            input_path = os.path.join(input_folder, file)
            output_path = os.path.join(output_folder, os.path.splitext(file)[0] + ".wav")
            
            audio = AudioSegment.from_file(input_path, format="m4a")
            audio.export(output_path, format="wav")
            print(f"Converted: {file} -> {output_path}")

# Example usage
input_folder = r"C:\Users\rohan\OneDrive\Documents\Gender detection\Vox_gender\females" # Change this to your input folder
output_folder = "wav_files"  # Change this to your output folder
convert_m4a_to_wav(input_folder, output_folder)
```
